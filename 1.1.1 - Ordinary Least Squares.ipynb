### 1.1 Linear Models
In mathematical notation, if $ \hat y $ is the predicted value,  $ \hat y(w,x) = w_0 + w_1x_1 + ... + w_px_p $. Across module 1 we designate the vector $w = (w_1,...,w_p)$ as coef_, and $w_0$ as intercept_

#### 1.1.1 Ordinary Least Squares
LinearRegression fits a linear model with coefficents $w = (w_1,...,w_p)$ to minimize the residual sum of squares between the observed targets in the dataset, and the targets predicted by the linear approximation. Mathematically it solves a problem of the form:
$$ \min_{w} = ||Xw - y||^2_2$$
This equation finds the value of w that minimizes $||Xw - y||^2_2$. $||Xw - y||^2_2$ represents the squared Euclidean norm of the difference between $Xw$ and $y$. 
* $X$ represents a matrix of input data, where each row is a data point and each column is a feature. 
* $w$ is a vector of weights or coefficients that we are trying to solve for. 
* $y$ is a vector of observed values or target values that we're trying to predict.



$||v||_2$ is just Euclidean distance

LinearRegression will take in its `fit` methods arrays `X` and `y`, and will store the coefficients $w$ of the linear model in its `coef_` member.
* `X` is a feature vector, an array of numbers that represent the properties or attributes of a single data sample. They are used by ML models to make predictions or decisions
* `y` is a target, the value or label that we want the model to predict based on those features

```python
from sklearn import linear_model

X = [[0, 0], [1, 1], [2, 2]]
y = [0, 1, 2]

reg = linear_model.LinearRegression()
reg.fit(X, y)
reg.coef_
```

This result should make sense, because $0(0.5) + 0(0.5) = 0$, $1(0.5) + 1(0.5) = 1$, and $2(0.5) + 2(0.5) = 2$

```python
# let's pretend X represents temperature, and y represents $ in ice cream sales

X = [[14.2], [16.4], [11.9], [15.2], [18.5], [22.1], [19.4], [25.1], [23.4], [18.1], [22.6], [17.2]]
y = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]

reg = linear_model.LinearRegression()
reg.fit(X, y)
print(reg.coef_)
print(reg.intercept_)
```

```python
# VISUALIZATION
import matplotlib.pyplot as plt

plt.scatter(X, y, color='black')
plt.plot(X, reg.predict(X), color='blue', linewidth=3)
plt.show()
```

#### 1.1.1.1 Non-Negative Least Squares
It's possible to constrain all the coefficients to be non-negative, which may be useful when they represent some physical or naturally non-negative quantities (e.g the prices of goods). LinearRegression accepts a 'positive' parameter. when set to True, non-negative least squares are then applied.

#### 1.1.1.2 Ordinary Least Squares Complexity
The least squares solution is computed using the singular value decomposition of X. If X is a matrix of shape (n_samples, n_features) this method has a cost of $O(n_{samples}\times n^2_{features})$, assuming that $n_{samples} \geq n_{features}$
